```python
import pandas as pd

distribution_centers = pd.read_csv('/content/distribution_centers.csv')
inventory = pd.read_csv('/content/inventory_items.csv')
```

```python
distribution_centers.head()
```

```python
inventory.head()
```

```python
inventory['created_at'] = pd.to_datetime(inventory['created_at'], errors='coerce')
inventory['sold_at'] = pd.to_datetime(inventory['sold_at'], errors='coerce')
```

```python
missing_values = inventory.isnull().sum()

print(inventory.head())
```

```python
inventory.dtypes
```

```python
distribution_centers = distribution_centers.drop_duplicates(subset=['name', 'latitude', 'longitude'])
```

```python
null_values = inventory.isnull().sum()

print("Null_values':")
print(null_values)
```

```python
inventory_distro = pd.merge(inventory, distribution_centers,
                             left_on=['product_distribution_center_id'],
                             right_on=['id'],
                             how='left',
                             suffixes=('_inv', '_dist'))

```

```python
inventory_distro = inventory_distro.drop(columns=['created_at_item_orders', "shipped_at_item_orders", "delivered_at_item_orders", "returned_at_item_orders"])
```

```python
inventory_distro.head()
```

```python
inventory_distro = inventory_distro.drop(columns=['id_dist'])
```

```python
inventory_distro.dtypes
```

```python
inventory_distro.to_csv('/content/inventory_distro.csv', index=False)
```
